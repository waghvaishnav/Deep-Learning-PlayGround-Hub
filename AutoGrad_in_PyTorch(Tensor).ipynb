## AutoGrad in PyTorch

Caculating Gradients using BackPropagation

```python
import torch
```

```python
material_cost = torch.tensor(10,requires_grad=True,dtype = torch.float16)
labor_cost = torch.tensor(35,requires_grad=True,dtype=torch.float16)

total_cost = 3 * material_cost**2 + 5 * labor_cost + 100
total_cost
```

```python
# apply backpropagation

total_cost.backward()


# returns gradients
material_cost.grad, labor_cost.grad
```

### Trying to passes a Vector

whwn we passes a vector they gives error for directly calculating grad, we have to sum of it then backward and calculate grad.

```python
x = torch.tensor([1,2,3],requires_grad=True,dtype=torch.float16)
y = 2 * x **3 + 7
y
```

```python
result = y.sum()
result.backward()
```

```python
x.grad
```
